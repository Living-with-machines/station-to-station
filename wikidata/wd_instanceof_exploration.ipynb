```python
import pandas as pd
import glob

from collections import Counter
from  itertools import chain
import pydash
import ast

from wikidata.client import Client
client = Client()

pd.set_option('display.max_columns', None)
```

### Explore "instance of" feature of entities

```python
britdf = pd.read_csv("britwikidata/british_isles.csv", header=0, index_col=None, low_memory=False)
```

```python
britdf = britdf.drop(columns=['Unnamed: 0'])
```

```python
britdf2 = britdf[~britdf.instance_of.isnull()]
```

```python
britdf2['instance_of'] = britdf2['instance_of'].apply(ast.literal_eval)
```

```python
britdf2.head()
```

```python
brit_instance = pd.Series(Counter(chain(*britdf2.instance_of))).sort_index()
```

```python
instance_of_frequency = brit_instance.to_dict()
```

```python
instance_of_label = dict()
```

```python
for k in instance_of_frequency:
    entity = client.get(k, load=True)
    record = entity.__dict__['data']
    instance_of_label[record['id']] = pydash.get(record, 'labels.en.value')
```

```python
len(instance_of_frequency)
```

```python
len(instance_of_label)
```

```python
# For some reason, some labels could not be retrieved, I add them by hand:
instance_of_label["Q18762207"] = "cemetery cross"
instance_of_label["Q87169767"] = "tunnel portal"
instance_of_label["Q17158079"] = "terminal train station"
```

```python
pd.DataFrame({'frequency':pd.Series(instance_of_frequency),'english_label':pd.Series(instance_of_label)}).to_csv("britwikidata/instance_of.csv")
```
