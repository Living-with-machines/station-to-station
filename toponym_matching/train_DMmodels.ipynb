# Training DeezyMatch models

```python
from DeezyMatch import train as dm_train
from DeezyMatch import plot_log
from DeezyMatch import finetune as dm_finetune
from DeezyMatch import inference as dm_inference

from pathlib import Path
from shutil import copyfile
```

## v001: char tokenizer

```python
if not Path('./models/wikigaz_en_001').is_dir():
    # train a new model
    dm_train(input_file_path="./inputs/input_dfm_001.yaml",
             dataset_path="./wikigaz_en_dataset.txt",
             model_name="wikigaz_en_001")
```

```python
# plot log file
plot_log(path2log="./models/wikigaz_en_001/log.txt", 
         output_name="wikigaz_en_001")
```

```python
best_checkpoint = "checkpoint00003"
```

```python
copyfile("./models/wikigaz_en_001/" + best_checkpoint + ".model", "./models/wikigaz_en_001/wikigaz_en_001.model")
copyfile("./models/wikigaz_en_001/" + best_checkpoint + ".model_state_dict", "./models/wikigaz_en_001/wikigaz_en_001.model_state_dict")
```
